```python
import numpy as np
import pandas as pd
import os
import re
```

```python
x = pd.DataFrame()
y = pd.Series()

for root, dirs, files in os.walk('tctodd'):
    for i,fn in enumerate(files):
        file = pd.DataFrame()
        if fn.endswith("tsd"):
            vals = pd.read_csv(os.path.join(root,fn),sep='\t',header=None)
            file = file.append(vals)
            y += pd.Series(re.search('(.+?)-[0-9]',fn).group(1))
        x = x.append(file)
```
